
# Learning Algorithm
The agent is using basic ddpg algorithm to solve the continuous control task. 
It uses a replay buffer to store past experience and uses batches of experiences to train the model.
This submission uses version 1 of the problem.

List of hyperparameters:

BUFFER_SIZE = int(1e6)  # replay buffer size  
BATCH_SIZE = 256        # minibatch size  
GAMMA = 0.9             # discount factor   
TAU = 1e-3              # for soft update of target parameters  
LR_ACTOR = 1e-3         # learning rate of the actor   
LR_CRITIC = 1e-3        # learning rate of the critic  
WEIGHT_DECAY = 0        # L2 weight decay  

In ddpg, the policy is determined by the actor networks and the quality of the policy is dertermined by the critic network.
In this perticular implementation the actor NN is made of 3 fully connected layers:   
The 1st layer has 33 inputs (the state space of the environment) and 128 outputs  
The 2nd layer has 128 inputs and 128 outputs  
The final layer has 128 inputs and 4 outputs with tanh activation (One for the torque on each joint) 

The critic NN is made of 3 fully connected layers as well: 
The 1st layer has 33 inputs (the state space of the environment) and 400 outputs  
The 2nd layer has 400+4 inputs (previous layer + outputs of actor network) and 300 outputs  
The final layer has 300 inputs and 1 outputs (The value of the policy) 

Agent was adapted from https://github.com/udacity/deep-reinforcement-learning/blob/master/ddpg-pendulum/ddpg_agent.py
